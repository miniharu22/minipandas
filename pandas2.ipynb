```python
df1.iloc[0:100]
# 데이터 추출 (행단위, iloc : index location)
    # 0번째부터 99번째까지 100개 데이터를 추출
    # head()와 유사함
```

```python
df1[['메세지','정지시간2','작업구분','작업자']]
# 데이터 추출 (열단위)
# Dataframe 형태로 데이터 추출
    # Series : index - value / []
    # Dataframe : index - column - value / [[]]
```

```python
cond1 = (df1['정지시간2'] >= 600)
df1.loc[cond1]
```

```python
cond1 = (df1['정지시간2'] >= 600)
cond2 = (df1['작업구분'] == '야간')

df1.loc[cond1 & cond2]
# 두 조건을 모두 만족하는 데이터만 추출
```

```python
cond1 = (df1['정지시간2'] >= 600)
cond2 = (df1['메세지'] == '초기화 Error')

df1.loc[cond1 | cond2]
# 두 조건 중 하나라도 만족하면 데이터를 추출
```

```python
df1.pivot_table(index='메세지',values='정지시간2',aggfunc= 'mean')
# 데이터 요약
```

```python
# 메세지 항목에서 'Door Open'에 해당하는 오류들 중에 '정지시간2'이 가장 오래 걸린 상위 100개 데이터를 추출
cond1 = (df1['메세지'] == 'Door Open')
df1.loc[cond1].sort_values(by = '정지시간2', ascending= False).head(100)
```

```python
# '정지시간2'가 300초 이상이면서, '작업자'가 '이은미'인 데이터를 추출해 '정지시간2'의 요약통계량을 확인
cond1 = (df1['정지시간2'] >= 300)
cond2 = (df1['작업자'] == '이은미')
df1.loc[cond1 & cond2].describe()
```
